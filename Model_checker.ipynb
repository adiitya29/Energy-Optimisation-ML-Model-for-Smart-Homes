loading the trained model in here

```python
import pickle

# Path to the model file in Google Drive
model_path = "/content/drive/My Drive/energy_optimizer_model.pkl"

# Load the trained model
with open(model_path, "rb") as file:
    loaded_model = pickle.load(file)

print("Model loaded successfully!")
```

loading and preprocessing of new dataset

```python
import pandas as pd

# Load the new dataset
df = pd.read_csv("/content/drive/My Drive/smart_home_energy_consumption_large.csv")

# Perform the same preprocessing steps as before
df['timestamp'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
df['hour'] = df['timestamp'].dt.hour
df['day'] = df['timestamp'].dt.day
df['month'] = df['timestamp'].dt.month
df.drop(columns=['Date', 'Time'], inplace=True)

# Drop target column (if present) to match training features
X_new = df.drop(columns=["Energy Consumption (kWh)"], errors="ignore")

print("Dataset preprocessed successfully!")
```

predictions using trained model

```python
from sklearn.preprocessing import LabelEncoder

# Ensure categorical columns are encoded the same way as in training
encoder_appliance = LabelEncoder()
encoder_season = LabelEncoder()

# Encode 'Appliance Type' and 'Season'
X_new['Appliance Type'] = encoder_appliance.fit_transform(X_new['Appliance Type'])
X_new['Season'] = encoder_season.fit_transform(X_new['Season'])

# Drop any unnecessary columns that were not part of training
X_new = X_new.drop(columns=['timestamp'], errors='ignore')

# Make predictions
predictions = loaded_model.predict(X_new)

# Add predictions to the dataset
df["Predicted Energy Consumption (kWh)"] = predictions

print(df.head())
```
